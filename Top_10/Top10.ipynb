```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files

# Upload your dataset
uploaded = files.upload()

# Load the dataset
df = pd.read_csv("Spotify_Youtube.csv")
```

#5. Top 10 song in spotify vs top 10 in youtube

```python
# Clean the data by removing rows with missing 'Stream' or 'Views'
df_clean = df.dropna(subset=["Stream", "Views"])

# Get top 10 songs by Spotify Streams
top10_spotify = df_clean.sort_values(by="Stream", ascending=False).head(10)

# Get top 10 songs by YouTube Views
top10_youtube = df_clean.sort_values(by="Views", ascending=False).head(10)

# Plot top 10 Spotify songs
plt.figure(figsize=(12, 6))
plt.barh(top10_spotify["Track"], top10_spotify["Stream"])
plt.xlabel("Spotify Streams")
plt.ylabel("Songs")
plt.title("Top 10 Songs on Spotify by Streams")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

# Plot top 10 YouTube songs
plt.figure(figsize=(12, 6))
plt.barh(top10_youtube["Track"], top10_youtube["Views"])
plt.xlabel("YouTube Views")
plt.ylabel("Songs")
plt.title("Top 10 Songs on YouTube by Views")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
